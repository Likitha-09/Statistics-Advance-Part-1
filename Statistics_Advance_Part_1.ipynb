1.What is a random variable in probability theory

In probability theory, a random variable is a function that assigns a numerical value to each possible outcome of a random experiment, allowing us to quantify and analyze these outcomes

2.What are the types of random variables

Random variables are classified into two main types: discrete and continuous, depending on whether they can take on a finite or infinite number of values, respectively.

3.What is the difference between discrete and continuous distributions

Discrete distributions deal with countable outcomes (like the number of heads in coin flips), while continuous distributions deal with outcomes that can take any value within a range (like height or weight).

4.What are probability distribution functions

 A probability distribution is a mathematical function that describes the probability of different possible values of a variable. Probability distributions are often depicted using graphs or probability tables

5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)

While both PDF (Probability Density Function) and CDF (Cumulative Distribution Function) describe the probability distribution of a random variable, the PDF focuses on the probability of a specific value, while the CDF provides the probability of a variable being less than or equal to a specific value.

6.What is a discrete uniform distribution

In statistics and probability theory, a discrete uniform distribution is a statistical distribution where the probability of outcomes is equally likely and with finite values.

7.What are the key properties of a Bernoulli distribution

1: The number of observations n is fixed.
2: Each observation is independent.
3: Each observation represents one of two outcomes ("success" or "failure").
4: The probability of "success" p is the same for each outcome.

8.What is the binomial distribution, and how is it used in probability

The binomial distribution is a discrete probability distribution that models the probability of a certain number of successes in a fixed number of independent trials, each with only two possible outcomes (success or failure), where the probability of success remains constant for each trial.

9.What is the Poisson distribution and where is it applied

A Poisson distribution is a discrete probability distribution. It gives the probability of an event happening a certain number of times (k) within a given interval of time or space. The Poisson distribution has only one parameter, λ (lambda), which is the mean number of events.

10.What is a continuous uniform distribution?

The uniform distribution (continuous) is one of the simplest probability distributions in statistics. It is a continuous distribution, this means that it takes values within a specified range, e.g. between 0 and 1.

11.What are the characteristics of a normal distribution?

Normal distributions are symmetric, unimodal, and asymptotic, and the mean, median, and mode are all equal. A normal distribution is perfectly symmetrical around its center. That is, the right side of the center is a mirror image of the left side.

12.What is the standard normal distribution, and why is it important?

The standard normal distribution is a specific type of normal distribution with a mean of 0 and a standard deviation of 1, often used as a reference for probability calculations and comparisons across datasets. It's important because it allows for easier probability calculations and comparisons across datasets with different means and standard deviations.

13.What is the Central Limit Theorem (CLT), and why is it critical in statistics

The Central Limit Theorem (CLT) states that the distribution of sample means will approximate a normal distribution as the sample size increases, regardless of the original population distribution. This is critical in statistics because it allows for robust statistical inference and hypothesis testing.


14.How does the Central Limit Theorem relate to the normal distribution

The Central Limit Theorem (CLT) is a cornerstone of statistics, stating that the distribution of sample means (the average of multiple samples) will approach a normal distribution, regardless of the original population's distribution, as the sample size increases.

15.What is the application of Z statistics in hypothesis testing.

Z statistics are used in hypothesis testing to determine if there's a statistically significant difference between a sample mean and a population mean (one-sample z-test) or to compare the means of two populations (two-sample z-test), when the population variance is known and the sample size is large.

16.How do you calculate a Z-score, and what does it represent

A Z-score, calculated as (x - μ) / σ, represents how many standard deviations a data point (x) is from the mean (μ) of a dataset, and is used to standardize data and compare values across different distributions.


17.What are point estimates and interval estimates in statistics

In statistics, a point estimate is a single value used to estimate an unknown population parameter, while an interval estimate provides a range of values within which the population parameter is likely to fall.

18.What is the significance of confidence intervals in statistical analysis

Confidence intervals in statistical analysis provide a range of plausible values for an unknown population parameter, offering a more informative and robust way to estimate parameters than relying solely on point estimates, while also quantifying the uncertainty inherent in sample-based estimates.

19.What is the relationship between a Z-score and a confidence interval

A Z-score, which represents the number of standard deviations a data point is from the mean, is crucial for calculating confidence intervals, as it determines the margin of error and the width of the interval.

20.How are Z-scores used to compare different distributions

Z-scores, also known as standard scores, standardize data by expressing a data point's distance from the mean in terms of standard deviations, allowing for comparisons across different distributions, even those with varying means and standard deviations.

21.What are the assumptions for applying the Central Limit Theorem

To apply the Central Limit Theorem (CLT), the samples must be independent, the sample size should be large (often considered 30 or more), and the population should have a finite variance.


22.What is the concept of expected value in a probability distribution?

In a probability distribution, the expected value (also known as the mean or mathematical expectation) is the weighted average of all possible outcomes, where each outcome is weighted by its probability of occurrence.

23.How does a probability distribution relate to the expected outcome of a random variable?

A probability distribution describes the likelihood of different outcomes for a random variable, and the expected outcome is calculated by weighting each possible value by its probability and summing these products.

```python
# Write a Python function to calculate the probability distribution function
 #(PDF) of a Bernoulli distribution>

def bernoulli_pdf(x, p):
    """
    Computes the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The value (0 or 1) for which the probability is calculated.
    p (float): The probability of success (1), where 0 <= p <= 1.

    Returns:
    float: The probability of obtaining x in a Bernoulli trial.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be in the range [0,1].")

    return p if x == 1 else (1 - p)

p_success = 0.7
print(bernoulli_pdf(1, p_success))
print(bernoulli_pdf(0, p_success))
```

```python
#Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_cdf(a, b):
    """
    Plots the cumulative distribution function (CDF) of a discrete uniform distribution
    defined over integers in the range [a, b].
    """
    x = np.arange(a, b+1)
    cdf_values = randint.cdf(x, a, b+1)

    plt.step(x, cdf_values, where='post', label=f'Discrete Uniform CDF [{a}, {b}]', linewidth=2)
    plt.xlabel('x')
    plt.ylabel('CDF')
    plt.title('Cumulative Distribution Function of Discrete Uniform Distribution')
    plt.xticks(x)
    plt.yticks(np.linspace(0, 1, b-a+2))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

a, b = 1, 10
plot_cdf(a, b)
```

```python
# Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores

def plot_z_scores(data):
    z_scores = calculate_z_scores(data)

    plt.figure(figsize=(10, 5))
    plt.scatter(range(len(data)), z_scores, color='b', label='Z-scores')
    plt.axhline(y=0, color='r', linestyle='--', label='Mean')
    plt.axhline(y=1.96, color='g', linestyle='--', label='95% CI Upper')
    plt.axhline(y=-1.96, color='g', linestyle='--', label='95% CI Lower')

    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.title('Z-score Plot')
    plt.legend()
    plt.grid()
    plt.show()


data = np.random.normal(50, 15, 100)
plot_z_scores(data)
```

```python
# Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def central_limit_theorem_demo(sample_size=30, num_samples=1000):
    """
    Demonstrates the Central Limit Theorem (CLT) using a non-normal distribution.

    Args:
    sample_size (int): Number of random values in each sample.
    num_samples (int): Number of samples to draw.
    """

    population = np.random.exponential(scale=2.0, size=100000)

    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

    plt.figure(figsize=(12, 6))
    sns.histplot(sample_means, bins=30, kde=True, color='blue', stat="density")
    plt.title(f'Central Limit Theorem Demonstration\nSample Size: {sample_size}, Number of Samples: {num_samples}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid()
    plt.show()


central_limit_theorem_demo()
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform Z-test for hypothesis testing.

    Parameters:
    sample : array-like
        The sample data.
    population_mean : float
        The mean of the population.
    population_std : float
        The standard deviation of the population.
    alpha : float, optional
        Significance level (default is 0.05).
    alternative : {'two-sided', 'less', 'greater'}, optional
        Defines the alternative hypothesis (default is 'two-sided').
        'two-sided' tests if the sample mean is different from population mean.
        'less' tests if the sample mean is less than the population mean.
        'greater' tests if the sample mean is greater than the population mean.

    Returns:
    tuple
        Z-score, p-value, and conclusion of hypothesis test.
    """
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_score)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("alternative hypothesis must be 'two-sided', 'less', or 'greater'")

    reject_null = p_value < alpha
    conclusion = "Reject the null hypothesis" if reject_null else "Fail to reject the null hypothesis"

    return z_score, p_value, conclusion

sample_data = [50, 52, 48, 49, 51, 53, 47, 50, 52, 51]
pop_mean = 50
pop_std = 3

z_score, p_value, result = z_test(sample_data, pop_mean, pop_std)
print(f"Z-score: {z_score:.2f}, P-value: {p_value:.4f}, Conclusion: {result}")
```

```python
#Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats


data = [88, 92, 85, 91, 89, 95, 90, 94, 87, 93]


mean = np.mean(data)
std_err = stats.sem(data)

confidence = 0.95
df = len(data) - 1
margin_of_error = stats.t.ppf((1 + confidence) / 2, df) * std_err

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")


```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats
np.random.seed(0)
data = np.random.normal(loc=10, scale=2, size=1000)

confidence_level = 0.95
mean = np.mean(data)
std_err = stats.sem(data)
h = std_err * stats.t.ppf((1 + confidence_level) / 2., len(data)-1)
conf_interval = (mean - h, mean + h)

print(f"Mean of the sample: {mean:.2f}")
print(f"Confidence interval for the mean (95%): {conf_interval}")

```

```python
#Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std_dev=1, size=1000):
    """
    Generates a dataset from a normal distribution and calculates its skewness and kurtosis.

    Parameters:
        mean (float): Mean of the normal distribution.
        std_dev (float): Standard deviation of the normal distribution.
        size (int): Number of samples to generate.

    Returns:
        tuple: (skewness, kurtosis)
    """
    data = np.random.normal(loc=mean, scale=std_dev, size=size)
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)

    return skewness, excess_kurtosis

skewness, kurt = calculate_skewness_kurtosis()
print(f"Skewness: {skewness}")
print(f"Excess Kurtosis: {kurt}")
```
